```
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
from sklearn.datasets import load_digits
```

```
data = load_digits()
x=data.data
y=data.target
```

```
kmeans=KMeans(n_clusters=10).fit_predict(x)
```

```
label=np.zeros_like(y)
```

```
from scipy import stats
```

```
for a in range(len(y)):
  label[a] = stats.mode(y[a])[0]
```

```
from sklearn.metrics import confusion_matrix
result=confusion_matrix(label,kmeans)
print(result)
```

```
import sklearn
sklearn.metrics.accuracy_score(label,kmeans)
```
