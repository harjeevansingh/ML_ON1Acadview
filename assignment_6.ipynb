```
##### Question 1 ####
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
df.head()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df.shape
```

```
col=list(df.columns)
col
```

```
itr =len(col)
```

```
#### countplot of Date
sns.countplot(x='Date', data=df)
sns.despine()
plt.title("1.Count plot of Date")
```

```
#### countplot of MinTemp
sns.countplot(x='MinTemp', data=df)
sns.despine()
plt.title("2.Count plot of MinTemp")
```

```
#### distplot of MaxTemp
sns.distplot(df['MaxTemp'])
sns.despine()
plt.title("3.Dist plot of MaxTemp")
```

```
#### distplot of Rainfall
sns.distplot(df['Rainfall'])
sns.despine()
plt.title("4.Dist plot of Rainfall")
```

```

sns.boxplot(x='Evaporation', y='Sunshine',data=df)
sns.despine()
plt.title("5.Box plot of 'Evaporation','Sunshine',")
```

```
sns.jointplot(x='Evaporation', y='Sunshine',data=df, alpha=0.2)
sns.despine()
plt.title("6.Dist plot of 'WindGustDir','WindGustSpeed'")
```

```
sns.boxplot(x= 'WindSpeed9am', y='WindSpeed3pm',data=df, hue='RainToday')
sns.despine()
plt.title("7.Dist plot of  'WindSpeed9am','WindSpeed3pm'")
```

```
sns.jointplot(x= 'Humidity9am', y='Humidity3pm',data=df)
sns.despine()
plt.title("8.Dist plot of  'Humidity9am','Humidity3pm'")
```

```
sns.violinplot(x= 'Pressure9am', y='RainTomorrow',data=df)
sns.despine()
plt.title("9.Dist plot of  'Pressure9am','RainTomorrow'")
```

```
df['RISK_MM'].plot(kind='barh')
sns.despine()
plt.title("10.Dist plot of MaxTemp")
```

```
sns.violinplot(x= 'Temp9am', y='Temp3pm',data=df,hue='RainToday' ,split= True, inner='stick',palette='Set2')
sns.despine()
plt.title("11.Dist plot of  'Temp9am','Temp3pm'")
```

```
sns.violinplot(x= 'RainToday', y='WindSpeed9am',data=df, hue='RainTomorrow',split=True,inner='stick')
sns.despine()
plt.title("12.Dist plot of  'RainToday','WindSpeed9am'")
```
